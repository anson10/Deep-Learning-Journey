```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs
```

```python
sns.scatterplot(data = blobs, x= "X1", y="X2")
```

```python
moons = pd.read_csv("cluster_moons.csv")
moons
```

```python
sns.scatterplot(data=moons, x='X1', y='X2')
```

```python
circles = pd.read_csv('cluster_circles.csv')
```

```python
sns.scatterplot(data=circles, x='X1', y='X2')
```

```python
def categories(model, data):
    labels = model.fit_predict(data)
    
    sns.scatterplot(data=data, x='X1', y='X2', hue=labels, palette='Set1')
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
```

```python
categories(model, blobs)
```

```python
model = KMeans(n_clusters=2)

categories(model, moons)
```

```python
categories(model, circles)
```

```python
from sklearn.cluster import DBSCAN

help(DBSCAN)
```

```python

model = DBSCAN()

categories(model, blobs)
```

```python
model = DBSCAN(eps=0.20)
categories(model, moons)
```

```python
categories(model, circles)
```
