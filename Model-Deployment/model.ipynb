```python
import pandas as pd 
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.describe()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
X_val, X__holdout_test, y_val, y_holdout_test = train_test_split(
    X_test, y_test, test_size=0.5, random_state=42
)
```

```python
len(X__holdout_test), len(X_val), len(X_train)
```

```python
# MODEL TRAINING
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=30, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

validation_prediction = model.predict(X_val)
```

```python
mean_absolute_error(y_val, validation_prediction)
```

```python
mean_absolute_error(y_val, validation_prediction)**0.5  # RMSE
```

```python
df.describe()['sales']
```

```python
# PERFORMANCE METRICS

holdout_predictions = model.predict(X__holdout_test)
```

```python
mean_absolute_error(y_holdout_test, holdout_predictions)
```

```python
mean_squared_error(y_holdout_test, holdout_predictions)**0.5
```

```python
final_model = RandomForestRegressor(n_estimators=30, random_state=42)
```

```python
final_model.fit(X, y)
```

```python
import joblib 
```

```python
joblib.dump(final_model, 'final_model.pkl')
```
